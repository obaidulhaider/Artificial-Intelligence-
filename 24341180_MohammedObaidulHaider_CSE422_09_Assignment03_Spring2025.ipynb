```python
#Problem1

import random
import math

def strength(x):
    return math.log(x + 1, 2) + (x / 10)

def utility(maxV, minV):

    i = random.randint(0, 1)
    random_factor = random.randint(1, 10)


    max_strength = strength(maxV)
    min_strength = strength(minV)
    return max_strength - min_strength + ((-1) ** i) * (random_factor / 10)

def minimax(depth, node, maximizing_player, alpha, beta, maxV, minV):
    if depth == 0:
        return utility(maxV, minV)

    if maximizing_player:
        best = float('-inf')
        for _ in range(2):
            best = max(best, minimax(depth - 1, node + 1, False, alpha, beta, maxV, minV))
            alpha = max(alpha, best)
            if beta <= alpha:
                break
        return best
    else:
        worst = float('inf')
        for _ in range(2):
            worst = min(worst, minimax(depth - 1, node + 1, True, alpha, beta, maxV, minV))
            beta = min(beta, worst)
            if beta <= alpha:
                break
        return worst


def determine_winner(utility_value, game, carlsen_strength, caruana_strength):

    if game % 2 != 0:
        if utility_value > 0:
            return "Magnus Carlsen (Max)", utility_value
        elif utility_value < 0:
            return "Fabiano Caruana (Min)", utility_value
        else:
            return "Draw", utility_value
    else:
        if utility_value > 0:
            return "Fabiano Caruana (Max)", utility_value
        elif utility_value < 0:
            return "Magnus Carlsen (Min)", utility_value
        else:
            return "Draw", utility_value

def start_game():

    first_player = int(input("Who goes first for game 1? (0 for Carlsen, 1 for Caruana): "))
    carlsen_strength = float(input("Enter Carlsen's base strength: "))
    caruana_strength = float(input("Enter Caruana's base strength: "))

    carlsen_wins = 0
    caruana_wins = 0
    draws = 0

    for game in range(1, 5):
        if game % 2 != 0:
            result = minimax(5, 0, True, float('-inf'), float('inf'), carlsen_strength, caruana_strength)
            winner, utility_value = determine_winner(result, game, carlsen_strength, caruana_strength)
            print(f"Game {game} Winner: {winner} (Utility value: {round(utility_value, 2)})")
        else:
            result = minimax(5, 0, True, float('-inf'), float('inf'), caruana_strength, carlsen_strength)
            winner, utility_value = determine_winner(result, game, carlsen_strength, caruana_strength)
            print(f"Game {game} Winner: {winner} (Utility value: {round(utility_value, 2)})")


        if winner == "Magnus Carlsen (Max)" or winner == "Magnus Carlsen (Min)":
            carlsen_wins += 1
        elif winner == "Fabiano Caruana (Max)" or winner == "Fabiano Caruana (Min)":
            caruana_wins += 1
        else:
            draws += 1


    print("\nOverall Results:")
    print(f"Magnus Carlsen Wins: {carlsen_wins}")
    print(f"Fabiano Caruana Wins: {caruana_wins}")
    print(f"Draws: {draws}")

    if carlsen_wins > caruana_wins:
        print("Overall Winner: Magnus Carlsen")
    elif caruana_wins > carlsen_wins:
        print("Overall Winner: Fabiano Caruana")
    else:
        print("Overall Winner: Draw")
start_game()
```

```python
#Problem2
def calculate_minimax_values(who_goes_first, cost, light_strength, l_strength):

    minimax_without_mc = (light_strength - l_strength) / (light_strength + l_strength)


    minimax_with_mc = 1.0


    minimax_with_mc_cost = minimax_with_mc - cost


    if minimax_without_mc > 0:

        use_mc = False
        reason = "Light should NOT use Mind Control as the position is already winning."
    elif minimax_with_mc_cost > minimax_without_mc:

        use_mc = True
        reason = "Light should use Mind Control."
    else:

        use_mc = False
        if minimax_with_mc_cost < minimax_without_mc:
            reason = "Light should NOT use Mind Control as it backfires."
        else:
            reason = "Light should NOT use Mind Control as the position is losing either way."


    return {
        "Minimax value without Mind Control": round(minimax_without_mc, 2),
        "Minimax value with Mind Control": round(minimax_with_mc, 2),
        "Minimax value with Mind Control after incurring the cost": round(minimax_with_mc_cost, 2),
        "Should Light use Mind Control?": use_mc,
        "Reason": reason
    }


who_goes_first = int(input("Enter who goes first (0 for Light, 1 for L): "))
cost = float(input("Enter the cost of using Mind Control: "))
light_strength = float(input("Enter base strength for Light: "))
l_strength = float(input("Enter base strength for L: "))

results = calculate_minimax_values(who_goes_first, cost, light_strength, l_strength)


print(f"Minimax value without Mind Control: {results['Minimax value without Mind Control']}")
print(f"Minimax value with Mind Control: {results['Minimax value with Mind Control']}")
print(f"Minimax value with Mind Control after incurring the cost: {results['Minimax value with Mind Control after incurring the cost']}")
print(results['Reason'])
```
